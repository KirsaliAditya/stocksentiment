Applying Sentimental Analysis

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
vader = SentimentIntensityAnalyzer()
```

```python
news['neg'] = 0.0
news['neu'] = 0.0
news['pos'] = 0.0
news['compound'] = 0.0

news
```

```python
#predicting summaries
def get_polarity_scores(text):
    return vader.polarity_scores(text)

for index, row in news.iterrows():
    scores = get_polarity_scores(row['summary'])
    news.at[index, 'neg'] = scores['neg']
    news.at[index, 'neu'] = scores['neu']
    news.at[index, 'pos'] = scores['pos']
    news.at[index, 'compound'] = scores['compound']
```

```python
# loading the saved model
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("/content/drive/MyDrive/finbert-fine-tuned")
model = AutoModelForSequenceClassification.from_pretrained("/content/drive/MyDrive/finbert-fine-tuned")
```

```python
# using the loaded model in news headlines title
import torch
def get_finbert_sentiment(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)
    outputs = model(**inputs)
    predicted_class = torch.argmax(outputs.logits, dim=1).item()
    label_map = {0: "negative", 1: "neutral", 2: "positive"}
    return label_map[predicted_class]
```

```python
news['headline_sentiment'] = news['title'].apply(get_finbert_sentiment)
```

```python
def combine_sentiments_weighted(row):
    # Map headline sentiment to numerical values
    sentiment_map = {"negative": -1, "neutral": 0, "positive": 1}
    headline_score = sentiment_map[row['headline_sentiment']]
    summary_score = row['compound']

    combined_score = 0.6 * headline_score + 0.4 * summary_score
    if combined_score > 0.05:
        return "positive"
    elif combined_score < -0.05:
        return "negative"
    else:
        return "neutral"

news['final_sentiment'] = news.apply(combine_sentiments_weighted, axis=1)
```

```python
import matplotlib.pyplot as plt

table_data = news[['title', 'headline_sentiment', 'compound', 'final_sentiment']].head(10)  # Adjust head(n) to limit rows displayed

fig, ax = plt.subplots(figsize=(10, 4))
ax.axis('tight')
ax.axis('off')

table = plt.table(
    cellText=table_data.values,
    colLabels=table_data.columns,
    cellLoc='center',
    loc='center',
    colColours=['#f2f2f2'] * len(table_data.columns)
)

table.auto_set_font_size(False)
table.set_fontsize(10)
table.auto_set_column_width(col=list(range(len(table_data.columns))))

plt.title("Visualization of Final Sentiment Table", fontsize=14)
plt.show()
```

```python
average_compound_score = news['compound'].mean()

vader_converted_score = ((average_compound_score + 1) / 2) * 100

finbert_sentiment_map = {"negative": -1, "neutral": 0, "positive": 1}
news['headline_sentiment_score'] = news['headline_sentiment'].map(finbert_sentiment_map)

average_headline_sentiment = news['headline_sentiment_score'].mean()
finbert_converted_score = ((average_headline_sentiment + 1) / 2) * 100

combined_converted_score = (
    0.4 * vader_converted_score + 0.6 * finbert_converted_score
)
```

```python
if combined_converted_score > 55:
    combined_sentiment = "Positive"
    recommendation = "Buy"
elif combined_converted_score < 45:
    combined_sentiment = "Negative"
    recommendation = "Sell"
else:
    combined_sentiment = "Neutral"
    recommendation = "Hold"

# Displaying results
print(f"Average VADER Sentiment Score (0-100): {vader_converted_score:.2f}")
print(f"Average FinBERT Sentiment Score (0-100): {finbert_converted_score:.2f}")
print(f"Aggregated Sentiment Score for the Stock (0-100): {combined_converted_score:.2f}")
print(f"Overall Sentiment: {combined_sentiment}")
print(f"Recommendation: {recommendation}")
```

```python
import matplotlib.pyplot as plt

# Data for visualization
scores = [vader_converted_score, finbert_converted_score, combined_converted_score]
labels = ["VADER Score", "FinBERT Score", "Aggregated Score"]
colors = ["skyblue", "lightgreen", "gold"]

# Define bar color based on recommendation
bar_color = "green" if combined_sentiment == "Positive" else "red" if combined_sentiment == "Negative" else "yellow"

# Plot bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, scores, color=colors, edgecolor="black", alpha=0.8)

# Highlight the aggregated score with the sentiment recommendation color
bars[2].set_color(bar_color)

# Add threshold lines
plt.axhline(y=55, color="green", linestyle="--", label="Positive Threshold (55)")
plt.axhline(y=45, color="red", linestyle="--", label="Negative Threshold (45)")

# Add labels and title
for i, score in enumerate(scores):
    plt.text(i, score + 2, f"{score:.2f}", ha="center", fontsize=12, weight="bold")

plt.title("Sentiment Score Breakdown and Recommendation", fontsize=14)
plt.ylabel("Sentiment Score (0-100)", fontsize=12)
plt.xticks(fontsize=10)
plt.ylim(0, 110)  # Increased ylim to make space for text above
plt.legend()

# Add final sentiment and recommendation as separate text boxes
plt.text(0.5, 105, f"Overall Sentiment: {combined_sentiment}", fontsize=12, color="black", bbox=dict(facecolor="lightgrey", alpha=0.7, pad=5))
plt.text(0.5, 95, f"Recommendation: {recommendation}", fontsize=12, color="black", bbox=dict(facecolor="lightgrey", alpha=0.7, pad=5))
# Show the plot
plt.tight_layout()
plt.show()
```
